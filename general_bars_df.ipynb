```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.graph_objects as go
from VersaQT.data_manip import *
```

```python
soy = yf.Ticker("ZS=F")
soy_history = soy.history(period="1y")
soy_history = soy_history.asfreq("B")

print(soy.info["shortName"])

# Interpolando os dados
soy_history = soy_history.interpolate()

soy_history.reset_index(inplace=True)

corn = yf.download("ZC=F", start="2020-01-01", end="2023-12-31", interval="1d")
corn.columns = corn.columns.get_level_values(0)
corn.reset_index(inplace=True)
corn
```

```python
tick_bar = getDollarBar(corn, 100000)
tick_bar

new_tick_bar = getBar(corn, 100000, "dollar", volume_col="Volume", dollar_col="Close")
```

```python

fig = go.Figure()

fig.add_trace(go.Scatter(x=tick_bar["StartTime"], y=tick_bar.Close, name="Tick bars"))
fig.add_trace(go.Scatter(x=new_tick_bar["StartTime"], y=new_tick_bar.Close, name="Volume Bars"))

fig.show()
```

```python
volume_bar = getVolumeBar(corn, 200_000)
volume_bar
```

```python
dollar_bar = getDollarBar(corn, 100_000_000_0)
dollar_bar
```

```python

fig = go.Figure()

fig.add_trace(go.Scatter(x=corn.Date, y=corn.Close, name="Time Series"))
fig.add_trace(go.Scatter(x=tick_bar["StartTime"], y=tick_bar.Close, name="Tick bars"))
fig.add_trace(go.Scatter(x=volume_bar["StartTime"], y=volume_bar.Close, name="Volume Bars"))
fig.add_trace(go.Scatter(x=dollar_bar["StartTime"], y=dollar_bar.Close, name="Dollar Bars"))

fig.show()
```
